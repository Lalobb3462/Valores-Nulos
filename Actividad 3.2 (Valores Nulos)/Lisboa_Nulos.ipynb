**Eliminaremos las columnas innecesarias para nuestro análisis**

**description**

```python
lisboa["description"]=lisboa["description"].fillna("Not available.")
#Hice uso de este método de sustitución por cadena ya que no se puede inventar una descripción por cada uno de los sitios que se conocen 
#y porque no podemos asignarle una descripción de otro lugar a ese. Con eso en mente, decidí mantener un mensaje que sea claro y nos diga cuales no tienen
#ni siquiera una descripción en su anuncio.
```

```python
nulos=lisboa.isnull().sum()
nulos
```

```python
lisboa["neighborhood_overview"]=lisboa["neighborhood_overview"].fillna("Not available")
#De la misma manera, al ser una descripción general del vecindario en que se encuentra, no podemos sobreponer datos ni generar uno para cada celda,
#por lo tanto opte de misma manera por mostrar con una cadena aquellos que no tengan este elemento.
```

```python
nulos=lisboa.isnull().sum()
nulos
```

**host_response_time**

**host_response_rate**

```python
lisboa["host_response_rate"]=lisboa["host_response_rate"].fillna("100%")
#En este caso le puse un valor de 100% ya que la mayoría de los datos en la tabla se acercan a estos valores, y por otro lado porque este valor es una
#constante que no pretende alterar los datos en gran escala.
```

**host_is_superhost**

```python
false=lisboa[lisboa["host_is_superhost"] == "f"]
false
#La mayoria no son superhost, es más probable que los nulos no lo sean.
```

```python
lisboa["host_is_superhost"]=lisboa["host_is_superhost"].fillna(method="bfill")
#En este caso, como no podemos comprobar que el anfitrión es un superhost, es mejor considerar que ninguno de los nulos lo es ya que no hay 
#una noción clara de quienes de ellos si lo pueden ser, pero como debo usar todos los métodos de sustitución optaré por usar el backward fill aquí,
#ya que al ser mayoría de valores falsos, tendera a tener más valores falsos incluso con ese método 
#(Comprobé y quedan el mismo número de falsos que si fuera una cadena f).
```

```python
nulos=lisboa.isnull().sum()
nulos
```

**bathrooms**

```python
lisboa["bathrooms"]=lisboa["bathrooms"].fillna(round(lisboaO["bathrooms"].mean()))
#Hice uso del promedio de todos los baños para mantenerlos en ese rango y no distorcionar la proporción de los baños totales en los distintos lugares.
```

```python
nulos=lisboa.isnull().sum()
nulos
```

**bedrooms**

```python
lisboa["bedrooms"]=lisboa["bedrooms"].fillna(round(lisboa["bedrooms"].median()))
#Hice uso de mediana ya que en las instrucciones era necesario ocupar cada método y considero que usarla en este ejemplo es buena opción.
```

```python
nulos=lisboa.isnull().sum()
nulos
```
